```python
import pandas as pd
import numpy as np
import featuretools as ft
```

```python
customer = pd.read_csv('/content/drive/MyDrive/SEM 5 ANIS/Customer.csv', encoding='ISO-8859-1')
product = pd.read_csv('/content/drive/MyDrive/SEM 5 ANIS/Product.csv', encoding='ISO-8859-1')
shipment = pd.read_csv('/content/drive/MyDrive/SEM 5 ANIS/Shipment.csv', encoding='ISO-8859-1')
payment = pd.read_csv('/content/drive/MyDrive/SEM 5 ANIS/Payment.csv', encoding='ISO-8859-1')
order = pd.read_csv('/content/drive/MyDrive/SEM 5 ANIS/orderNew.csv', encoding='ISO-8859-1')
```

```python
# Ensure that the OrderDate column is in datetime format with dayfirst=True
order['OrderDate'] = pd.to_datetime(order['OrderDate'], dayfirst=True, errors='coerce')
```

```python
# Ensure that the SignUpDate column is in datetime format with dayfirst=True
customer['SignupDate'] = pd.to_datetime(customer['SignupDate'], dayfirst=True, errors='coerce')
```

```python
# Ensure that the ShipDate column is in datetime format with dayfirst=True
shipment['ShipDate'] = pd.to_datetime(shipment['ShipDate'], dayfirst=True, errors='coerce')
```

Check for duplicated in PK column

```python
print(customer['CustomerID'].duplicated().any())
print(product['ProductID'].duplicated().any())
print(order['OrderID'].duplicated().any())
print(payment['PaymentID'].duplicated().any())
print(shipment['ShipID'].duplicated().any())
```

```python
# Drop duplicates in primary key column
customer = customer.drop_duplicates(subset = ['CustomerID'])

print(customer['CustomerID'].duplicated().any())
```

```python
# Drop duplicates in primary key column
product = product.drop_duplicates(subset = ['ProductID'])

print(product['ProductID'].duplicated().any())
```

```python
# Ensure all missing values are dropped
df_list = {
    'Customer': customer,
    'Order': order,
    'Product': product,
    'Payment': payment,
    'Shipment': shipment
}
null = []

for name, df in df_list.items():
    # Count missing values in each column
    null.append(df.isna().sum())

# Print null content
for name, null_values in zip(df_list.keys(), null):
    print(f"Null values in {name} DataFrame:")
    print(null_values)
    print("\n")
```

**Define Entities and EntitySet**

```python
# creating and entity set
es = ft.EntitySet(id = 'ecommerce_store')
```

Define entities

```python
# Adding a DataFrame as an entity
es.add_dataframe(dataframe_name='customer_entity', dataframe=customer, index='CustomerID')
es.add_dataframe(dataframe_name='product_entity', dataframe=product, index='ProductID')
es.add_dataframe(dataframe_name='shipment_entity', dataframe=shipment, index='ShipID')
es.add_dataframe(dataframe_name='order_entity', dataframe=order, index='OrderID')
es.add_dataframe(dataframe_name='payment_entity', dataframe=payment, index='PaymentID')
```

make relationship

```python
es.add_relationship(parent_dataframe_name='shipment_entity', parent_column_name='ShipID',
                    child_dataframe_name='order_entity', child_column_name='ShipID')
es.add_relationship(parent_dataframe_name='customer_entity', parent_column_name='CustomerID',
                    child_dataframe_name='order_entity', child_column_name='CustomerID')
es.add_relationship(parent_dataframe_name='product_entity', parent_column_name='ProductID',
                    child_dataframe_name='order_entity', child_column_name='ProductID')
es.add_relationship(parent_dataframe_name='payment_entity', parent_column_name='PaymentID',
                    child_dataframe_name='order_entity', child_column_name='PaymentID')
```

Perform Deep Feauture Synthesis

```python
# !pip install dask
```

```python
feature_matrix, feature_names = ft.dfs(entityset=es,
                                       target_dataframe_name = 'order_entity')
```
